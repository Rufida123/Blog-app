```python
import pandas as pd
```

```python
path="/content/student_lifestyle_dataset.csv"
data = pd.read_csv(path)
data .head(10)
```

```python
print(data.shape)
data.info()
```

```python
# التحقق من القيم المفقودة
print("القيم المفقودة في كل عمود:")
print(data.isnull().sum())
print("شكل البيانات بعد التنظيف:", data.shape)
```

```python
# عرض أسماء جميع الأعمدة
print("أسماء الأعمدة المتاحة:")
for col in data.columns:
    print(f"- {col}")
```

```python
features = ['Study_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day']
target = 'GPA'
print("المميزات المختارة:", features)
print("الهدف:", target)
```

```python
# مجرد عرض للعلاقات بين المميزات والهدف
import matplotlib.pyplot as plt

# هذه مجرد رسوم بيانية بسيطة لفهم البيانات
data[features + [target]].hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
# العلاقة بين المميزات والهدف
correlations = data[features + [target]].corr()
print("العلاقات بين المتغيرات:")
print(correlations[target].sort_values(ascending=False))
```
